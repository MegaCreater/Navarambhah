```python
import cv2




# Opencv DNN
net = cv2.dnn.readNet("dnn_model/yolov4-tiny.weights", "dnn_model/yolov4-tiny.cfg")
model = cv2.dnn_DetectionModel(net)
model.setInputParams(size=(320, 320), scale=1/255)

# Load class lists
classes = []
with open("dnn_model/classes.txt", "r") as file_object:
    for class_name in file_object.readlines():
        class_name = class_name.strip()
        classes.append(class_name)

print("Objects list")
print(classes)
KNOWN_WIDTH = 20  # Width of the object in real-world units (inches, centimeters, etc.)
FOCAL_LENGTH = 100 


    # Initialize camera
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)
# FULL HD 1920 x 1080'''

cv2.namedWindow("Frame")


# Create window

while True:
    # Get frames
    ret, frame = cap.read()

    # Get active buttons list
    #
    #print("Active buttons", active_buttons)

    # Object Detection
    (class_ids, scores, bboxes) = model.detect(frame)
    for class_id, score, bbox in zip(class_ids, scores, bboxes):
        (x,y,w,h) = bbox
        class_name=classes[class_id]
        cv2.putText(frame,(class_name),(x,y-5),cv2.FONT_HERSHEY_PLAIN,2,(255,0,50),2)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,50),3)
        distance = (KNOWN_WIDTH * FOCAL_LENGTH) / w

        text = f" {distance:.2f} units"  # Modify 'units' to match your real-world units
        cv2.putText(frame, text, (x, y - 20), cv2.FONT_HERSHEY_PLAIN, 0.5, (255, 0, 0), 2)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        
        print("Class ids", class_id)
        print("Scores", score)
        print("Bboxes", bbox)
        
        

        


    # Display buttons

    cv2.imshow("Frame", frame)
    key = cv2.waitKey(1)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
